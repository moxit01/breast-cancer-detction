```python
y = df['diagnosis']
y.head()
```

```python
x = df.drop('diagnosis', axis =1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

```python
df.shape
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty='l2', random_state=51)
lr.fit(x_train, y_train)
```

```python
y_pred_lr = lr.predict(x_test)
```

```python
y_pred_lr
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_lr)
accuracy
```

```python
results =pd.DataFrame()
results
```

```python
tempResults =pd.DataFrame({'Algorithm': ['Logistic Regression'], 'Accuracy':[accuracy]})
results =pd.concat([results, tempResults])
results =results[['Algorithm', 'Accuracy']]
results
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)
```

```python
y_pred_dt = dtc.predict(x_test)
y_pred_dt
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_dt)
accuracy
```

```python
tempResults = pd.DataFrame({'Algorithm':['Decision tree Classifier Method'], 'Accuracy':[accuracy]})
results = pd.concat( [results, tempResults] )
results = results[['Algorithm','Accuracy']]
results
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
```

```python
y_pred_rfc = rfc.predict(x_test)
y_pred_rfc
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_rfc)
accuracy
```

```python
tempResults =pd.DataFrame({'Algorithm': ['Random Forest'], 'Accuracy':[accuracy]})
results =pd.concat([results, tempResults])
results =results[['Algorithm', 'Accuracy']]
results
```

```python
from sklearn import svm
svc = svm.SVC()
svc.fit(x_train,y_train)
```

```python
y_pred_svc = svc.predict(x_test)
y_pred_svc
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred_svc)
accuracy
```

```python
tempResults =pd.DataFrame({'Algorithm': ['SVC'], 'Accuracy':[accuracy]})
results =pd.concat([results, tempResults])
results =results[['Algorithm', 'Accuracy']]
results
```

```python
from xgboost import XGBClassifier
xgb_classifier = XGBClassifier()
xgb_classifier.fit(x_train, y_train)
y_pred_xgb = xgb_classifier.predict(x_test)
accuracy = accuracy_score(y_test, y_pred_xgb)
accuracy
```
